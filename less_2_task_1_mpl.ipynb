```python
fig_1, AX = plt.subplots(2,2, gridspec_kw = {'hspace' : 0.3, 'wspace': 0.3}, figsize = (8,6))

AX[0,0].plot(x,y1)
AX[0,0].set(xlim = (-5,5))

AX[0,1].plot(x,y2)

AX[1,0].plot(x,y3)

AX[1,1].plot(x,y4)


n = 1
for ax in AX.flat:
    ax.set_title( f'График y{n}', fontsize = 7)
    n+=1
```
